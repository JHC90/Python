```python
# Bibs
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# load data
advertising=pd.read_csv('./Sample-Projects/ISLR/data/islrData_advertising.csv',delimiter=',',encoding='utf-8')
advertising = advertising.drop(['Unnamed: 0'], axis=1)
```

## Grundlagen Scatterplot

### Single-Plot

```python
sns.scatterplot(x=advertising.TV, y=advertising.sales, data=advertising) # so wird die grafik direkt ausgegeben
var = sns.scatterplot(x=advertising.TV, y=advertising.sales, data=advertising) # jetzt ist die Grafik in der Varibale Test
# hinterlegt und kann dann bspw in mehreren Plots dargesteltl werden
```

### Multiplot

```python
cols = ['TV','radio','newspaper']
fig,axes = plt.subplots(1,3,figsize=(20,6))
for seg,col in enumerate(cols):
    x,y = seg//3,seg%3
    an=sns.scatterplot(x=col, y='sales' ,data=advertising, ax=axes[y])
    plt.setp(an.get_xticklabels(), rotation=45)
```
