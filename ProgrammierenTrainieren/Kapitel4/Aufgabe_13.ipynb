```python
def testpalindrom(testzahl):
    convertToString = str(testzahl)
    lenght = len(convertToString)
    
    # erst die Geraden Fälle
    if(len(convertToString) % 2 == 0):
        testlength = int(len(convertToString)/2)
        for i in range(0,testlength,1):
            if(convertToString[0+i] != convertToString[-(1+i)]):
                return(False)
            else:
                pass
        return(True)
        
    else:
        testlength = int(round(len(convertToString)/2)) # mittlere Zahl
        for i in range(0,testlength,1):
            if(convertToString[0+i] != convertToString[-(1+i)]):
                return(False)
            else:
                pass
        return(True)
        
```

```python
testpalindrom(testzahl1)
```
