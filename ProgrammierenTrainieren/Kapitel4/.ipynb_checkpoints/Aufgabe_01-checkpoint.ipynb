```python
# hier mit variabler anzahl an incoming argumenten
def bestimmeMaximum(*incoming):  
    maxValue = 0
    for i in range(0, len(incoming[0]), 1):
        if incoming[0][i] <= maxValue:
                pass
        else:
            maxValue = incoming[0][i]
            
    return maxValue
```

```python
testinglist1 = [1, 2, 3] # hier die liste erstellen
testinglist2 = [1, 2, 3,4, 5, 6] # hier die liste erstellen
testinglist3 = [1456, 2891632, 3987654321,4234, 54564, 6123123]
```

```python
bestimmeMaximum(testinglist1)
```

```python
bestimmeMaximum(testinglist2)
```

```python
bestimmeMaximum(testinglist3)
```
