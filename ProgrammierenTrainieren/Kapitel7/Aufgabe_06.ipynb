```python
def checkBrackets(incoming):
    counter = 0 # counter open Brackets
    print(incoming)
    for i in range(0, len(incoming)):
        
       
        # bei geöffneter breket eine hinzufügen 
        if(incoming[i] == "("):
            counter += 1
            continue
        # bei geschlossener Klammer eins raus
        elif(incoming[i] == ")"):
            counter -= 1
            
            # prüfen ob der Counter jetzt negativ durch die Verringerung
            if(counter < 0):
                print("Schließende Klammer an dem Index: " + str(i) + " wurde noch nicht geöffnet")
                break
            continue
    if(counter > 0):
        print("es wurden mehr Klammern geöffnet als geschlossen")
    elif(counter == 0):
        print("Sollte alles passen")
    elif(counter < 0):
        print("Fehlerfal")
    else:
        print("fehler in der else Verzweigung => speziell nochmals nachschauen. Konkret geht es um den Wert" + str(incoming))
        
    
       
    print(counter)
    print("#######")
        
        
        
        
```

```python
for i in range(0, len(testvalues),1):
    checkBrackets(testvalues[i])
```
