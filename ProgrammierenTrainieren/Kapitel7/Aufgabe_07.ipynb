```python
def printStar(incoming):
    
    print()
    print()
    print("###############- "+ str(incoming)+" -##############")
    
    #oberer Teil
    for i in range(1, incoming,1):
        print("*"*i)
    # zentraler teil
    print("*"*(incoming*2-1))
    
    #endingpart
    leadingZeros = (" "*incoming)
    #enumeratingZeros = "" # den Raus um auf die Musterlösung zu kommen 
    for i in range(incoming-1, 0, -1):

        #print(str(leadingZeros) + str(enumeratingZeros) + str("*"*i))
        print(str(leadingZeros) + str("*"*i))
        #enumeratingZeros += " " # den Raus um auf die Musterlösung zu kommen 
    print("##################################")
    print("##################################")
    print("##################################")
```

```python
# Hier der Testlauf über den Loop
for i in range(0,len(testvalues),1):
    printStar(testvalues[i])
    
```
