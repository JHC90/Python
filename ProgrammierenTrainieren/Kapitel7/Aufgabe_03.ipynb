```python
alphaNumDic ={
    
    "a" : 1,
    "b" : 2, 
    "c" : 3, 
    "d" : 4, 
    "e" : 5, 
    "f" : 6, 
    "g" : 7,
    "h" : 8, 
    "i" : 9,
    "j" : 10, 
    "k" : 11, 
    "l" : 12, 
    "m" : 13, 
    "n" : 14, 
    "o" : 15, 
    "p" : 16, 
    "q" : 17, 
    "r" : 18, 
    "s" : 19, 
    "t" : 20, 
    "u" : 21, 
    "v" : 22, 
    "w" : 23, 
    "x" : 24, 
    "y" : 25, 
    "z" : 26
}
```

```python
def calcPrfZiffer(incoming):
    calcVector = []
    interimNumber= str((alphaNumDic[incoming[0].lower()]))
    for i in range(0,len(interimNumber),1):
        calcVector.append(interimNumber[i])
    for j in range(1,len(incoming) - 1,1):
        calcVector.append(incoming[j])
    
    
    # quersumme 
    quersumme = 0
    for ziffer in calcVector:
        quersumme += int(ziffer)
    
    
    # ganzzahliger Rest = Modulo
    # minuend - subtrahend = differenz
    
    subtrahend = quersumme % 9
    
    pruefzifferCalc = 8 - subtrahend # hier die errechnung
    pruefzifferVal = 0 # hier erst mal dummy, dass ich die Variable habe, die Prfe ich dann gegen die letzte Ziffer
    if(pruefzifferCalc == 0):
        pruefzifferVal = 9
    else:
        pruefzifferVal = pruefzifferCalc
        
    # prfen ob sich die Richtige Prfziffer gegebn hat
    if(int(incoming[len(incoming)-1]) == int(pruefzifferVal)):
        return("keine Flschung")
    else:
        return("Blte")
        
        
   
        
        
    
        
   
```

```python
calcPrfZiffer(teststring1)
```

```python
calcPrfZiffer(teststring2)
```

```python
calcPrfZiffer(teststring3)
```
