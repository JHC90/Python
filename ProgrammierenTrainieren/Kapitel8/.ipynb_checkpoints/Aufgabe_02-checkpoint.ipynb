```python
teststring = "ab;cde;fghi;jklm" #→ ["ab", "cde", "fghi", "jklm"]
```

```python
import re

def splitString(incoming):
    x = re.split("\;", incoming)
    print(x)
```

```python
print(teststring)
splitString(teststring)
```
