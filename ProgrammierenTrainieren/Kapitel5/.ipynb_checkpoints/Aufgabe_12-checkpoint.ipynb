```python
def generateSandClock(maxValue):
    iinterim = maxValue
    spaceinterim = ""
    while(iinterim >= 1):
        currentlineinterim = spaceinterim + "#" * iinterim 
        print(currentlineinterim)
        spaceinterim = spaceinterim + " "
        iinterim = iinterim -2 
    iinterim = iinterim + 2  
    spaceinterim = spaceinterim[1:]
    while(iinterim != maxValue+2):
        currentlineinterim = spaceinterim + "#" * iinterim 
        print(currentlineinterim)
        spaceinterim = spaceinterim[1:]
        iinterim = iinterim + 2 
    
```

```python
generateSandClock(50)
```
