```python
def checkLeapYear(yearTBC):
    if(yearTBC % 4 == 0):
        if(yearTBC % 100 == 0 and yearTBC % 400 != 0):
            print(str(yearTBC) + " ist kein Schaltjahr")
        else:
            print(str(yearTBC) + " ist ein Schaltjahr")
    else:
        print(str(yearTBC) + " kann kein Schaltjahr sein, da es ungerade ist")         
```

```python
checkLeapYear(2012)
```

```python
checkLeapYear(2000)
```

```python
checkLeapYear(1916)
```
