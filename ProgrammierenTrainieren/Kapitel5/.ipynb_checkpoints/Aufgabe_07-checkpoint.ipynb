```python
def calcDegreeAnalogHour(h,m):
    return(0.5*(60*h+m))
```

```python
def calcDegreeAnalogMin(m):
    return(6*m)
```

```python
import re
def checkTime(strIncoming):
    x = re.split("\:", strIncoming)
    degreeHour = calcDegreeAnalogHour(int(x[0]), int(x[1]))
    print("Stundenzeiger an der Stunde: " + x[0] + 
          "\nund der Minute: " + x[1] + 
          "\nhat den Winkel: " + str(degreeHour))
    print("Der Minutenanzeiger hat hierbei den Winkel: " + str(calcDegreeAnalogMin(int(x[1]))))
```

```python
checkTime("11:15")
```

```python
checkTime("3:33")
```
