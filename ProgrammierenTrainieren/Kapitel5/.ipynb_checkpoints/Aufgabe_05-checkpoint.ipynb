```python
import re
def volumenConverter(incoming):
    # typumwanglund regex nur string
    strIncoming = str(incoming)
    # convert to a string without signs
    x = re.split("\.", strIncoming)
    workingstring = str(x[0]) +str(x[1])
    if(workingstring[0] != "0"):
        return(strIncoming +" Liter")
    elif(workingstring[1] != "0"):
        print(strIncoming)
        return(strIncoming[2:4] + "0 cl(10)")
    elif(workingstring[2] != "0"):
        return(strIncoming[3:] +" cl(1)")
    elif(workingstring[3] != "0"):
        return(strIncoming[4:] + " ml")
    else:
        return("Nicht erfasst")
    return(workingstring)

    
   
    
        
```

```python
volumenConverter(3.0)
```

```python
volumenConverter(0.30)
```

```python
print(volumenConverter(0.03))
```
