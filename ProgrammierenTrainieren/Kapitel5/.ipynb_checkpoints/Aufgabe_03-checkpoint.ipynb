```python
def computeGarmenSize(sex, hight, chestsize):
    print(sex)
    print(type(sex))
    if sex == "m":
        confectionSize = chestsize / 2
        return(confectionSize)
    else:
        confectionSizeNormalsize = (chestsize / 2) - 6
        if(hight > 164 and hight < 170):
            normalSize = confectionSizeNormalsize
            return(normalSize)
        elif(hight<164):
            shortSize = confectionSizeNormalsize / 2
            return(shortSize)
        elif(hight>170):
            largeSize = confectionSizeNormalsize * 2
            return(largeSize)
        else:
            pass
        
           
        
```

```python
computeGarmenSize("m", 192, 92)
```
