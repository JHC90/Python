```python
def generateIban(kntnr, blz):
    ziffernString = str(blz) + str(kntnr) + "131400"
    pruefsummeinterim = int(ziffernString) % 97
    showdown = str(98 -pruefsummeinterim)
   
    
    # auf stelle 2 erweitern
    if(len(showdown) > 1):
        prueffinal = showdown
    else:
        prueffinal = "0" + str(showdown)
    
    return("DE" + prueffinal + str(blz) + str(kntnr)) 
```

```python
generateIban(1234567890, 70090100)
```
