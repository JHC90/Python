```python
import math
# Berechnung grad in Bogenmaß
def calcRadian(degreeIn):
    radian = ((degreeIn/180)*math.pi)
    return(radian)
```

```python
def calcDistance(lat1, lon1, lat2, lon2):
    '''
    print((calcRadian(lat1)))
    print((calcRadian(lon1)))
    print((calcRadian(lat2)))
    print((calcRadian(lon2)))
    '''
    c = ((math.sin(calcRadian(lat1)) *
           math.sin(calcRadian(lat2))
           ) + 
          (math.cos(calcRadian(lat1)) *
           math.cos(calcRadian(lat2)) * 
           math.cos(calcRadian(lon2) - calcRadian(lon1))
          )
          )

    print(c)
    print(math.acos(c))
    d = math.acos(c) * 6378,137   
    return(d)
```

```python
calcDistance(50.094157, 6.95821, 51.21795, 6.76165)
```
