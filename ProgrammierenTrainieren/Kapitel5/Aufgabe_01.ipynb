```python
def doWork(e, s):
    if(s <= 0 or e <= 0 ):
        return(0)
    if(e < s):
        eInt = e
        e = s
        s = eInt
        
    result = 1
    for i in range(s, e+1,1):
        result *= i
        
    return(result)
```

```python
doWork(3,4)
```

```python
doWork(5,2)
```

```python
doWork(-3,5)
```

```python
doWork(33,0)
```

```python
doWork(0,33)
```
