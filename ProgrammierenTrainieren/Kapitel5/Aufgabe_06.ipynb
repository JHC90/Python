```python
# Create Dictionary
# Schadstoffklasse 3-Achs >4 Achs
dictionarySchadstoffklasse = {
    "A": [12.5, 13.1],
    "B": [14.6, 15.2],
    "C": [15.7, 16.3],
    "D": [18.8, 19.4],
    "E": [19.8, 20.4],
    "F": [20.8, 21.4]
}
```

```python
def calcMaut(schadStKl, achsenN, kmN, currentDict):
    currentPricingModel = currentDict[schadStKl]
    if achsenN <= 3:
        return(currentPricingModel[0] * kmN)
    else:
        return(currentPricingModel[1] * kmN)
    return(currentPricingModel)
    '''
    print(schadStKl)
    print(type(schadStKl))
    print(achsenN)
    print(type(achsenN))
    print(kmN)
    print(type(kmN))
    print(currentDict)
    '''
    
    
```
