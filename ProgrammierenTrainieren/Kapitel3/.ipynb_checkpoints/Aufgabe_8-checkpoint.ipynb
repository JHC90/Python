```python
def createGameField(hight, wide):
    field = (4*wide+1)*"+"
    field = field +"\n"
    for i in range (0, hight,1): # loop ber die Hhe
        for j in range(0,wide,1):
            field = field + "|   " # pip & 3x leer
        field = field + "|\n" #new line
        field = field + wide * "++++"
        field = field + "+\n"
    
    return(field)
    
    

    
```

```python
print(createGameField(7,5))
```
