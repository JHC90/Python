```python
import math
pi = math.pi
```

```python
def calcVolumen(length,diameter):
    volume = length*(diameter*diameter*(pi/4))
    return(volume)
```

```python
print(calcVolumen(10, 0.30))
```

```python
print(calcVolumen(15, 0.32))
```
