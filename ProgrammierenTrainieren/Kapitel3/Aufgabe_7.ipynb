```python
def recalc(R, G, B):
    
    # w und wmax ermitteln
    W = [R/255, G/255, B/255]
    W = max(W)
    
    # c
    C = ((W-R/255)/W)
    
    # M
    M = ((W-G/255)/W)
    # Y
    Y = ((W-B/255)/W)
    # K
    K = 1 - W
    
    return(C, M, Y, K)
   

    
```

```python
incoming = (recalc(75, 0, 130))
print("an der Stelle c: " + str(incoming[0]))
print("an der Stelle y: " + str(incoming[1]))
print("an der Stelle m: " + str(incoming[2]))
print("an der Stelle k: " + str(incoming[3]))
```
