```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pi = np.pi
R = 5 # Radius
n = 80 # number of edges // je mehr desto glatter der Kreis // bsp 4 gibt 4-eck zurück, je mehr desto länger die Rechnung
t = np.linspace(0, 2*np.pi, n+1)
x = R*np.cos(t)
y = R*np.sin(t)
```

```python
plt.axis("equal")
plt.grid() # Gitternetz
plt.plot(x,y)
plt.show()
```

```python
r2 = 7
n2 = 80 # number of edges // je mehr desto glatter der Kreis // bsp 4 gibt 4-eck zurück, je mehr desto länger die Rechnung
t2 = np.linspace(0, 2*np.pi+5, n2+1)
x2 = r2*np.cos(t2) + 5
y2 = r2*np.sin(t2) + 5 # mit +- int kann ich in y verschieben
```

```python
plt.axis("equal")
plt.grid() # Gitternetz
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```
