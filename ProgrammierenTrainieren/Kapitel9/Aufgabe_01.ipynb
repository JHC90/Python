```python
teststring1 = "WWWWBBBWBBBBBBWW" #  "4W3B1W6B2W"
teststring2 = "BBBBWWWWWWWWWB" #  "4B9W1B"
teststring3 = "WBBBBWWWWWWB" #  "1W4B6W1B"
```

```python
def compress(incoming):
    returnvalue = ""
    currentChar = incoming[0]
    laufvariable = 1
    
    for i in range (1,len(incoming),1):
        if(incoming[i] == currentChar):
            laufvariable += 1
        else:
            returnvalue = returnvalue + str(laufvariable) +str(currentChar)
            currentChar = incoming[i]
            laufvariable = 1
    returnvalue = returnvalue + str(laufvariable) +str(currentChar)
    return(returnvalue)
        
    
    
```

```python
compressedString = compress(teststring1)
print(compressedString)
```

```python
compressedString = compress(teststring2)
print(compressedString)
```

```python
compressedString = compress(teststring3)
print(compressedString)
```
