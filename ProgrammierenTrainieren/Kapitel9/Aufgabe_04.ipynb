```python
testdict =[
    
    "eVJo2!8IrRo",
    "aH6*LauTp21u",
    "S3cr3ts!",
    "Passwort123", # => kein Sonderzeichen
    "!2Bcv", # => zu Kurz
    "123456", # => nur Zahlen
    "1234567", # => nur Zahlen
    "12345678" # => nur Zahlen
    
]
```

```python
import re
def checkRequirements(incoming):
    print(incoming)
    
    # Lnge
    if(len(incoming)<8):
        print("passwort ist zu Kurz")
    else:
        print("Passwortlnge passt")
     
    # Kleinbuchstaben
    if(bool(re.search('[a-z]', incoming))):
        print("Kleinbuchstaben sind drin")
    else:
        print("Kleinbuchstaben FEHLEN")
        
     # GroBuchstaben
    if(bool(re.search('[A-Z]', incoming))):
        print("Grobuchstaben sind drin")
    else:
        print("Kleinbuchstaben FEHLEN")
        
     # Ziffern
    if(bool(re.search('[0-9]', incoming))):
        print("Zahlen sind drin")
    else:
        print("Zahlen FEHLEN")
    
    # Sonderzeichen 
    # // einfach alle Buchstab und Zahlen negieren => sonderzeichensatz
    if(bool(re.search('[^A-Za-z0-9\_]', incoming))):
        print("Sonderzeichen sind drin")
    else:
        print("Sonderzeichen FEHLEN")
        
    
```

```python
for entry in testdict:
    checkRequirements(entry)
    print("###---NEW---###\n")
    
```
