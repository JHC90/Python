```python
X = [0, 1, 2, 3, 4, 5, 6, 7]  # für das etwaige Plotten die x-Werte
F = [2, 2, 4, 6, 10, 6, 4, 2] # Values zum Glaetten
Z = [2, 2, 4, 6, 7, 6, 4, 3] # Ziel aus der Angabe....so soll F aussehen wenn dir Funktion, die ich schreibe fertig ist
```

```python
def glaetten(incoming):
    returnlist = []
    for i in range(0,len(incoming),1):
        if(i == 0):
            returnlist.append((incoming[i]+incoming[i+1])/2)
        elif(i == len(incoming)-1):
            returnlist.append((incoming[-1] + incoming[-2]) / 2)
            return(returnlist)
            break
        else:
            returnlist.append(int((incoming[i-1]+incoming[i]+incoming[i+1])/3))
    print(returnlist)
```

```python
geglaetteteArray = glaetten(F)
print(geglaetteteArray)
# Array ist geglaette nach wunsch und Vorgabe => 

```

```python
import matplotlib.pyplot as plt

plt.plot(X,F,geglaetteteArray)
#plt.show() # das nur wenn wir im raw coding und nicht im Notebook sind
```
