```python
# for plotting the Picture
def printarray(incoming):
    LengthLevel1 = len(incoming) # nur die Anzahl der Zeilen
    LengthLevel2 = len(incoming[1]) # nur die Anzahl der spalten
    emptyString = ""
    for i in range(0, LengthLevel1,1): # Zeile
        for j in range(0, LengthLevel2,1): #Spalten
            if(incoming[i][j] == 1):
                emptyString = emptyString + "|#|"
            else:
                emptyString = emptyString + "| |"
        emptyString = emptyString + "\n"
    print(emptyString)
                   
            
            
        
    
```

```python
def createnewArray(incomingOriginalArray, factor):
    incomingLength = len(incomingOriginalArray)
    incomingcolum = len(incomingOriginalArray[0])
    # first Vector
    currentMatrix = []
    for i in range(0,incomingLength,1): # zeilen iteration
        currentVector = [] # muss in jeder Zeile der ausgangs/basisarray erstellt werden
        for j in range(0, incomingcolum,1): # Spalten-Iterator jeder wert der basis array itteriert
            # mit i und j habe ich auf jeden wert der basis array  zugegriffenm jetzt kommt der tatsäliche zoom
            for k in range(0,factor,1): 
                currentVector.append(incomingOriginalArray[i][j])
        # den currentvecotr "n"/ "faktor" mal in die Matrix hängen
        #print(currentVector)
        #print(type(currentVector))
        for k in range(0,factor,1):
            currentMatrix.append(currentVector)
    return(currentMatrix) 
    
    
    
    
  
   
```
