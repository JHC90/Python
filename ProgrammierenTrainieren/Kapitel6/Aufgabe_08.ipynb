```python
import math
testingArray1 = [4, 8, 6, 1, 2, 9, 4, 3]
testingArray2 = [33,8,1,6,43,54]


def checkMinDist(incoming):
    dist  = math.sqrt((incoming[0]-incoming[1])**2)
    IndexTBR = ""

    for i in range(0,len(incoming)-1,1):
        if((math.sqrt((incoming[i]-incoming[i+1])**2))< dist):
            dist = (math.sqrt((incoming[i]-incoming[i+1])**2))
            IndexTBR = i
        else:
            pass
    return(dist, IndexTBR)




```
