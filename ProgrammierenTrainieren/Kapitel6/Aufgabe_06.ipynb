```python
def printArena(currentArena):
    for i in range(0, len(currentArena),1):
        printrow = []
        for j in range(0, len(currentArena[i]),1):
            if(currentArena[i][j] == 1):
                printrow.append("#")
            else:
                printrow.append("_")
        print(printrow)
```

```python
from random import *

def createArena(incoming):
    currentSeatsinArena =[]
    for i in range(incoming, 2, -1):
        # Aktuelle besetzung
        CurrentSeats = []
        for j in range(0,i,1):
            CurrentSeats.append(randint(1,2))
        currentSeatsinArena.append(CurrentSeats)
    return(currentSeatsinArena)
  
```

```python
currentArenaPlan = (createArena(12))
printArena(currentArenaPlan)
```
