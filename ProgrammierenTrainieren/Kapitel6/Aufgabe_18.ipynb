```python
import math

def crossCalc(incoming, ReturnValueIncoming):
    # Zhlvariable fr die Ausgabe in den Returnvalue, sprich dass ich mich immer wieder auf die Orginal Array beziehen kann
    # Returnvalue

    currentTestingPoint = incoming[0]
    currentCrossValList = incoming[1:]
    #print(len(currentCrossValList))
    if(len(currentCrossValList) == 0):
        pass
    else:
        for i in range(0, len(currentCrossValList),1):
            # hier die Calc
            print#(currentTestingPoint[0])
            print#(currentTestingPoint[1])
            print#(currentCrossValList[i][0])
            print#(currentCrossValList[i][1])
            dist = math.sqrt((currentTestingPoint[0]-currentCrossValList[i][0]) ** 2 +
                             (currentTestingPoint[1] - currentCrossValList[i][1])**2)
            #print(dist)
            ReturnValueIncoming.append([currentTestingPoint[0],currentTestingPoint[1],
                                        currentCrossValList[i][0],currentCrossValList[i][1], 
                                        dist])
            
        crossCalc(currentCrossValList, ReturnValueIncoming)
        

```
