```python
def checkLeapYear(yearTBC):
    if(yearTBC % 4 == 0):
        if(yearTBC % 100 == 0 and yearTBC % 400 != 0): #F
            return(False)
        else:#T
            return(True)
    else:#F
        return(False)      
```

```python
import re
def CheckWeekday(incoming):
    
    # split in Time-Values
    x = re.split("\.", incoming) 
    day = x[0]
    month = x[1]
    year =x [2]
    
    # 1St Value mit nur der ersten Zahl
    value1Interim = str(int(year[2:])/4)[0]
    value2Interim = int(year[2:]) + int(value1Interim)
    print(value2Interim)
    
    # Addition abhängig vom Monat
    monthDictionarry = {"01":1,
                     "02":4,
                     "03":4,
                     "04":0,
                     "05":3,
                        "06":5,
                        "07":0,
                        "08":3,
                        "09":6,
                        "10":1,
                        "12":6,
                     "11":4
                    }
    value2Interim = value2Interim +monthDictionarry[month]
    print(value2Interim)
    
    # Value 3 interim
    value3interim = value2Interim + int(day)
    print(value3interim)
    
    # Jahrhunder
    centuryDictionarry = {
        "18":2,
        "19":0,
        "20":6,
        "21":4
    }
    value4interim = value3interim + centuryDictionarry[str(year)[:2]]
    print("value4interim: " +str(value4interim))
    
    # check schaltjahr and switch jan & feb
    if(checkLeapYear(int(year))):
        print("Schaltjahr")
        if(month == "01"):
            value5interim = value4interim -1
        elif(month == "02"):
            value5interim = value4interim -1
        else:
            value5interim = value4interim
        
    else:
        print("Kein Schaltjahr")
        value5interim = value4interim
        
    # check weekday
    
    centuryDictionarry = {
         0:"Sa",
         1:"So",
         2:"Mo",
         3:"Di",
         4:"Mi",
         5:"Do",
         6:"Fr"
    }
    print(value5interim)
    print(value5interim % 7)
    print(centuryDictionarry[value5interim % 7])
    
    
```

```python
CheckWeekday("13.11.2013")
```
