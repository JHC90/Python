```python
from random import *
def printfield(fieldnumber):
    #prüfung ob gerade oder nicht, is mir jetzt wurscht
    anzahlPaare = (fieldnumber ** 2)/2

    sortedGameDeck = []
    for i in range(0, int(anzahlPaare),1):
        for j in range(0,2,1):
            sortedGameDeck.append(i)

    # deck mixen
    mixedGameDeck = []
    for i in range(0,fieldnumber,1): # zeilenerstellung
        mixedGameRow = []
        for j in range(0,fieldnumber,1):# Spaltenerstellung in Zeile
            randomfromSortedDeck = randint(0, len(sortedGameDeck)-1)
            mixedGameRow.append(sortedGameDeck[randomfromSortedDeck])
            sortedGameDeck.pop(randomfromSortedDeck)

        mixedGameDeck.append(mixedGameRow)
    return(mixedGameDeck)
```

```python
currentGameField = printfield(4)
for i in range(0,5,1):
    print("############## || Beispiel: " + str(i+1) + " || ##############")
    
    for i in range(0,len(currentGameField),1):
        currentGameField = printfield(4)
        print(currentGameField[i])
```
