```python
exampleSolution = [
    [4,1,7,9,2,5,8,3,6],
    [3,2,9,6,7,8,1,5,4],
    [6,8,5,3,4,1,7,9,2],
    
    [2,9,1,5,6,4,3,7,8],
    [5,6,3,2,8,7,9,4,1],
    [8,7,4,1,3,9,2,6,5],
    
    [9,4,2,7,1,6,5,8,3],
    [1,5,8,4,9,3,6,2,7],
    [7,3,6,8,5,2,4,1,9]
    
]
```

```python
# Check current TestingArray from Row, Colum, Block
def checkTesting(incoming):
    for i in range(0, len(incoming),1):
        testAgainstVector = list(range(1,10))
        checkvalue = "t" # fr den Fall eines treffers muss man abbrechen
        for j in range(0, len(incoming[0]),1):
            if((incoming[i][j]) in testAgainstVector):
                testAgainstVector.pop(testAgainstVector.index(incoming[i][j]))
            else:
                checkvalue = "f"
                print("nein")
                break
        if(checkvalue == "f"):
            print(incoming[i])
            print("breakvalue in zeile " +str(i+1) + " und Spalte "+str(j+1) +
                  " im TestingVector// nicht notgedrungen in der TestingMatrix")
            break
```

```python
# Row CHECK
def createRowTesting(incoming):
    testingarray=[]
    for i in range(0,len(incoming),1):
        testingarray.append(incoming[i])
    return(testingarray)

# hier der call
rowsToBeChecked = createRowTesting(exampleSolution)
checkTesting(rowsToBeChecked)
    
```

```python
# Block CHECK
def createBlockTesting(incoming):
    # erstelle die Blcke, zeilen sprung 3
    testingarray=[]
    for i in range(0,7,3): # Zeilenstpnge
         for j in range(0,7,3): # spaltensprnge
                #print(i, j) # jedes linke bere Eck
                
                # aufspanne der submatrizen
                currentBlock = []
                for k in range(0,3,1):
                    for l in range(0,3,1):
                        currentBlock.append(incoming[i+k][j+l])
                # hinzufgen den Blocks in das testing
                testingarray.append(currentBlock)
       
    return(testingarray)

blocksToBeChecked = createBlockTesting(exampleSolution)
checkTesting(blocksToBeChecked)
```

```python
# Column CHECK
def createColumnTesting(incoming):
    testingarray=[]
    for i in range(0,len(incoming),1): #spalteniterator
        # current AppendingVector under Construction
        currentAppendVector = []
        for j in range(0,len(incoming[i]),1): # zeilenIteratot
            currentAppendVector.append(incoming[j][i])
        testingarray.append(currentAppendVector)
    return(testingarray)


# hier der call
columnsToBeChecked = createColumnTesting(exampleSolution)
checkTesting(columnsToBeChecked)
       
    
```
