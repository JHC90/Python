```python
# for plotting the Picture
def printarray(incoming):
    LengthLevel1 = len(incoming) # nur die Anzahl der Zeilen
    LengthLevel2 = len(incoming[1]) # nur die Anzahl der spalten
    emptyString = ""
    for i in range(0, LengthLevel1,1): # Zeile
        for j in range(0, LengthLevel2,1): #Spalten
            if(incoming[i][j] == 0):
                emptyString = emptyString + "| |"
            elif(incoming[i][j] == 1):
                emptyString = emptyString + "|#|"
            elif(incoming[i][j] == 2):
                emptyString = emptyString + "|~|"
            elif(incoming[i][j] == 3):
                emptyString = emptyString + "|%|"
            elif(incoming[i][j] == 4):
                emptyString = emptyString + "|$|"
            elif(incoming[i][j] == 5):
                emptyString = emptyString + "|&|"
            else:
                pass
        emptyString = emptyString + "\n"
    print(emptyString)
                   
            
            
        
    
```

```python
basicarray3= [[2,0,5,0,0],
              [3,4,1,0,0],
              [4,1,0,0,2],
              [3,2,3,0,0]
]

def frame(incoming):
    returnValue = []
    
    pictureWithFrame = []
    createdTopAndEndVector = [1] * (len(incoming[0])+2)
    iterimVec = []
    for i in range(0, len(incoming)):
        
        # Build same vector as in all rows
        interimVec = [1] + incoming[i] + [1]      
        
        if(i == 0): # erste Zeile = oben Bildleiste
            pictureWithFrame.append(createdTopAndEndVector)
            pictureWithFrame.append(interimVec)
        elif(i == len(incoming)-1): # letzte Zeile = unten Bildleiste
            pictureWithFrame.append(interimVec)
            pictureWithFrame.append(createdTopAndEndVector)
        else:
            pictureWithFrame.append(interimVec)
    return(pictureWithFrame)
           
            
```

```python
framedPic = frame(basicarray3)
```

```python
for i in range(0, len(basicarray3),1):
    print(basicarray3[i])
print("##### => ")
for i in range(0, len(framedPic),1):
    print(framedPic[i])
print("##### => ")
printarray(framedPic)
```
