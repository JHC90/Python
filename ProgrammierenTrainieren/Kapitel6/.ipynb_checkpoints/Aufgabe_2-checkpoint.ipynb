```python
testarray = [123,134,120,122]
```

```python
def calcMean(incoming):
    print(len(incoming))
    totaldistance = 0
    for i in range(0, len(incoming), 1):
        totaldistance += incoming[i]
    return(totaldistance/len(incoming))
```

```python
print(calcMean(testarray))
```
