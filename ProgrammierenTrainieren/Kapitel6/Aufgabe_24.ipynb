```python
from random import *
def createTestPicture(incomingRows,incomingColumns):
    
    randomPicture = []
    for i in range(0,incomingRows,1):
        currentRowVectorUnderConstruction = []
        for j in range(0,incomingColumns,1):
            currentRowVectorUnderConstruction.append(randint(0, 256))
        randomPicture.append(currentRowVectorUnderConstruction)
    return(randomPicture)
        
```

```python
def createErrorsInPicture(incoming):
    minZeilenWert = 1
    maxZeilenWert = len(incoming) -2
    minSpaltenWert = 1
    maxSpatenWert = len(incoming[0]) -2
    # insgesamt felder =
    anzahlFelder = len(incoming) * len(incoming[0])
    anzahlRauschen = int(0.1*anzahlFelder)
    
    # rauschen einbauen
    for i in range(0,anzahlRauschen,1):
        #random x
        randX = randint(minZeilenWert, maxZeilenWert)
        randY = randint(minSpaltenWert, maxSpatenWert)
        incoming[randX][randY] = 255
    return(incoming)
        
    
    
    
```

```python
picWithNoError = createTestPicture(8,12)
```

```python
picWithErrors = createErrorsInPicture(picWithNoError)

for i in range(0,len(picWithErrors),1):
    print(picWithErrors[i])
    #print(len(picWithErrors[i]))
```

# hier die Korrektur-Function

```python
# Hier die Korrekto
def correct(incoming):
    
    for i in range(1, len(incoming)-1,1):
        for j in range(1,len(incoming[0])-1,1):
            if((incoming[i][j]) == 255):
                
                # wenn treffer dann spann folgene Matrix auf und sortiere sie
                currentBlock = [] 
                for k in range(-1,2,1): # currentBlock rows
                    for l in range(-1,2,1):
                        #print(incoming[i+k][j+l])
                        currentBlock.append(incoming[i+k][j+l])
                currentBlock.sort()
                #print(currentBlock)
                #print(currentBlock[4])
                #print(currentBlock[int(len(incoming)/2)])
                incoming[i][j] = currentBlock[int(len(incoming)/2)]
    return(incoming)
                
               
                
            
        
        
```

```python
## immer das ganze file durchlaufen lassen, es bringt nicht die Korrektur von der Korrektur zu machen

for i in range(0, len(picWithErrors),1):
    print(picWithErrors[i])
print("###################")

correctedPic = correct(picWithErrors)

for i in range(0, len(correctedPic),1):
    print(correctedPic[i])
print("###################")
```
