```python
import re
def checkRequirements(incoming):
    if(bool(re.search('^[\S+]+@+[\S+]+[\.]+[\S+]{2,3}$', incoming))):
        print("Email Passt")
    else:
        print("Email Passt Nicht")
    
    
    
      
        
    
```

```python
for entry in testdict:
    checkRequirements(entry)
    print("###---NEW---###\n")
    
```
